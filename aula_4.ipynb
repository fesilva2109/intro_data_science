```python
# prompt: consegue consertar esse codigo?

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carrega o conjunto de dados (substitua 'tmdb_5000_movies.csv' pelo caminho real)
tmdb = pd.read_csv("tmdb_5000_movies.csv")

# Exibe as primeiras linhas do DataFrame
print(tmdb.head())

# Conta a frequncia de filmes em idiomas diferentes do ingls
total_outros_filmes = tmdb.query("original_language != 'en'")["original_language"].value_counts()

# Cria um grfico de barras para mostrar a distribuio de filmes em outros idiomas
plt.figure(figsize=(12, 6))
sns.countplot(
    data=tmdb.query("original_language != 'en'"),
    palette="mako",
    x="original_language",
    order=total_outros_filmes.index,
)
plt.xticks(rotation=90)
plt.title("Nmero de Filmes em Idiomas Diferentes do Ingls")
plt.xlabel("Idioma Original")
plt.ylabel("Contagem")
plt.show()

# Cria um grfico de barras para mostrar a porcentagem de filmes em outros idiomas
plt.figure(figsize=(12, 6))
sns.countplot(
    data=tmdb.query("original_language != 'en'"),
    palette="mako",
    x="original_language",
    order=total_outros_filmes.index,
    hue="original_language",
    hue_order=total_outros_filmes.index,
    stat="percent",
)
plt.xticks(rotation=90)
plt.title("Porcentagem de Filmes em Idiomas Diferentes do Ingls")
plt.xlabel("Idioma Original")
plt.ylabel("Porcentagem")
plt.legend(title="Idioma Original", loc="upper right")
plt.show()

# Cria exemplos de avaliaes de filmes
filme1 = [2.5] * 10 + [3.5] * 10
filme2 = [5] * 10 + [1] * 10
filme0 = [3.0] * 10

# Calcula a mdia e a mediana das avaliaes
media_filme1 = np.mean(filme1)
mediana_filme1 = np.median(filme1)

media_filme2 = np.mean(filme2)
mediana_filme2 = np.median(filme2)

media_filme0 = np.mean(filme0)

# Exibe os resultados
print("\nFilme 1:")
print("Mdia:", media_filme1)
print("Mediana:", mediana_filme1)

print("\nFilme 2:")
print("Mdia:", media_filme2)
print("Mediana:", mediana_filme2)

print("\nFilme 0:")
print("Mdia:", media_filme0)

# Calcula o desvio padro das avaliaes
desvio_padrao_filme0 = np.std(filme0)
desvio_padrao_filme1 = np.std(filme1)
desvio_padrao_filme2 = np.std(filme2)

# Exibe os resultados
print("\nDesvio Padro:")
print("Filme 0:", desvio_padrao_filme0)
print("Filme 1:", desvio_padrao_filme1)
print("Filme 2:", desvio_padrao_filme2)
```
