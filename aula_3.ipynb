```python
tmdb["original_language"].unique()
```

```python
contagem_lingua = tmdb["original_language"].value_counts().to_frame().reset_index()
contagem_lingua.columns = ["original_language","total"]
contagem_lingua.head()
```

```python
sns.barplot(x="original_language", y="total", data=contagem_lingua)
plt.title("Language Distribution")
plt.xlabel("Language")
plt.ylabel("Total")
plt.show()
```

```python
contagem_lingua.plot(kind="pie",y="total",labels = contagem_lingua["original_language"])
plt.title("Language Distribution")
```

```python
total_lingua = tmdb["original_language"].value_counts()
total_geral = total_lingua.sum()
total_ingles = total_lingua.loc["en"]
total_resto = total_geral- total_ingles
print(total_geral,total_ingles, total_resto)
```

```python
sns.barplot(x=["ingles","Outras LÃ­nguas"], y=[total_ingles,total_resto])
plt.title("Language Distribution")
plt.xlabel("Language")
plt.ylabel("Total")
plt.show()
```

```python
dados = {
    "lingua" : ["InglÃªs","Outras LÃ­nguas"],
    "total" : [total_ingles,total_resto]
}
dados = pd.DataFrame(dados)
dados
```

```python
dados.plot(kind="pie",y="total",labels = dados["lingua"])
plt.title("Language Distribution")
```

```python
sns.barplot(data = dados, x = "lingua", y = "total")
plt.title("Language Distribution")
plt.xlabel("Language")
```
